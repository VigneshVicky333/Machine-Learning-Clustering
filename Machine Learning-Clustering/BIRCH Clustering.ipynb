```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import Birch
from sklearn.preprocessing import StandardScaler
```

```python
# Load the dataset
data = pd.read_csv('Mall_Customers.csv')
```

```python
# Preview the dataset
print(data.head())
```

```python
# Select relevant features (Annual Income and Spending Score)
X = data[['Annual Income (k$)', 'Spending Score (1-100)']].values
```

```python
# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Create and fit the BIRCH model
birch = Birch(n_clusters=5)  # Specify the number of clusters
labels = birch.fit_predict(X_scaled)
```

```python
# Add the labels to the original data
data['Cluster'] = labels
```

```python
# Print the results
print("Labels:", labels)
print("Number of clusters found:", len(set(labels)) - (1 if -1 in labels else 0))  # Excluding noise
```

```python
# Plot the results
plt.figure(figsize=(10, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels, cmap='viridis', marker='o', s=50)
plt.title("BIRCH Clustering of Mall Customers")
plt.xlabel("Standardized Annual Income")
plt.ylabel("Standardized Spending Score")
plt.colorbar(label='Cluster Label')
plt.show()
```
