```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import SpectralClustering
from sklearn.preprocessing import StandardScaler
```

```python
# Load the dataset
data = pd.read_csv('Mall_Customers.csv')
```

```python
# Preview the dataset
print(data.head())
```

```python
# Select relevant features (Annual Income and Spending Score)
X = data[['Annual Income (k$)', 'Spending Score (1-100)']].values
```

```python
# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Create and fit the Spectral Clustering model
spectral_clustering = SpectralClustering(n_clusters=5, affinity='nearest_neighbors', random_state=0)
labels = spectral_clustering.fit_predict(X_scaled)
```

```python
# Print the results
print("Labels:", labels)
```

```python
# Plot the results
plt.figure(figsize=(10, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels, cmap='viridis', marker='o')
plt.title("Spectral Clustering of Mall Customers")
plt.xlabel("Standardized Annual Income")
plt.ylabel("Standardized Spending Score")
plt.show()
```
